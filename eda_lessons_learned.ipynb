```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 300

data = {
    "date": pd.date_range(start="2022-01-01", periods=n, freq="D"),
    "category": np.random.choice(
        ["Training", "Equipment", "Personnel", "Logistics", "Communications"], n
    ),
    "severity": np.random.randint(1, 6, n),
    "unit_type": np.random.choice(
        ["Company", "Battalion", "Brigade"], n
    ),
    "status": np.random.choice(
        ["Open", "Closed"], n, p=[0.3, 0.7]
    )
}

df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("lessons_learned_sample.csv", index=False)
```
